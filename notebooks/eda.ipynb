### Análisis Univariante

```python
cols_cat = data.select_dtypes(include="bool").columns
print(cols_cat)
print("\nCantidad de categorias:", len(cols_cat))
```

```python
fig, ax = plt.subplots(4, 3, figsize=(12, 10))
fig.subplots_adjust(hspace=0.3)
ax = ax.flatten()

for i, col in enumerate(cols_cat):
    ax[i].pie(data[col].value_counts(), labels=data[col].unique(), autopct='%1.1f%%')
    ax[i].set_title(col)
```

Observamos el gran desbalance existente en las categorías.

**Nota:** Hay que tener en cuenta que *isHomophobic* e *isRadicalism* son el 100% False y que *IsSexist*, *IsNationalist* e *IsReligiousHate* tienen más del 98% False.

Se eliminan las features *IsHomophobic* e *IsRadicalism* ya que no aportaran valor al modelo, puesto que solo tienen un unico valor.

```python
data = data.drop(["IsHomophobic", "IsRadicalism"], axis=1)
print("Cols:", data.columns)

cols_cat = data.select_dtypes(include="bool").columns
```

#### Multi clase o Multi etiqueta

```python
# Verificar si es multietiqueta o multiclase

label_counts = data[cols_cat].sum(axis=1)
is_multi_label = (label_counts > 1).any()
print("Es multi etiqueta?: ", is_multi_label)
```

El dataset es multiequeta

```python
# label_counts[label_counts >= 1].count()
label_counts[label_counts >= 1].count()
label_counts[label_counts >= 1]
label_counts >= 1
```

### Analizando columna Text

```python
from wordcloud import WordCloud, STOPWORDS
```

#### Palabras más frecuentes de todos los comentarios.

```python
text_list = data["Text"].to_list()
wordcloud = WordCloud(stopwords=STOPWORDS).generate(" ".join(text_list))

plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

#### Palabras más frecuentes por cada categoría

```python
comm_toxic = data[data["IsThreat"] == True]
comm_toxic = comm_toxic["Text"].to_list()

wordcloud2 = WordCloud().generate(" ".join(comm_toxic))
plt.imshow(wordcloud2, interpolation='bilinear')
plt.axis("off")
```
